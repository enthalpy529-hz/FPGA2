```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
edges = cv2.Canny(gray,50,150,apertureSize = 3)
plt.imshow(edges)
```

```python
lines = cv2.HoughLines(edges,1,np.pi/180,40)
```

```python
for line in lines:
    print(str(line[0,0])+' '+str(line[0,1]))
```

```python
# 



# 

pic2 = np.ones(np.shape(img))*255
# 
num = 70
lines = cv2.HoughLines(edges,1,np.pi/180,num)

# 
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv2.line(pic2,(x1,y1),(x2,y2),(0,0,255),2)

# 
cv2.imshow('img',pic2)
cv2.waitKey(0)
#cv2.destroyAllWindows()
cv2.imwrite(str(num)+'.jpg', pic2)
```
